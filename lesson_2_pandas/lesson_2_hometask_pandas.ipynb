```python
import pandas as pd
import numpy as np
import re
```

```python
def answer_one():
    energy = pd.read_excel('Energy Indicators.xls', skiprows=17, skipfooter=38, usecols="C:F", 
                           names=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable'])
    
    pd.set_option('future.no_silent_downcasting', True)
    energy = energy.replace('...', np.nan).infer_objects(copy=False)
    energy['Energy Supply'] *= 1000000
    
    def clean_country(name):
        name = re.sub(r'\d+', '', name)
        name = re.sub(r' \(.*\)', '', name)
        return name.strip()
    
    energy['Country'] = energy['Country'].apply(clean_country)
    energy['Country'] = energy['Country'].replace({
        "Republic of Korea": "South Korea",
        "United States of America": "United States",
        "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
        "China, Hong Kong Special Administrative Region": "Hong Kong"
    })
    GDP = pd.read_excel('world_bank.xls', skiprows=3) 
    GDP['Country Name'] = GDP['Country Name'].replace({
        "Korea, Rep.": "South Korea",
        "Iran, Islamic Rep.": "Iran",
        "Hong Kong SAR, China": "Hong Kong"
    })
    GDP.rename(columns={'Country Name': 'Country'}, inplace=True)
    
    ScimEn = pd.read_excel('scimagojr.xlsx')
    ScimEn_top15 = ScimEn[:15]
    
    df = pd.merge(ScimEn_top15, energy, how='inner', on='Country')
    
    columns_to_keep = ['Country', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    df = pd.merge(df, GDP[columns_to_keep], how='inner', on='Country')
    
    df.set_index('Country', inplace=True)

    df.drop('Region', axis=1, inplace=True)
    
    return df
```

```python
def answer_two():
    Top15 = answer_one()
    years = ['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    avgGDP = Top15[years].mean(axis=1, skipna=True).sort_values(ascending=False)
    avgGDP.name = 'avgGDP'
    return avgGDP
```

```python
answer_two()
```

```python
def answer_three():
    Top15 = answer_one()
    avgGDP = answer_two()
    country_6th = avgGDP.index[5]
    gdp_change = Top15.loc[country_6th, '2015'] - Top15.loc[country_6th, '2006']
    return gdp_change
```

```python
answer_three()
```

```python
def answer_four():
    Top15 = answer_one()
    Top15['Citation Ratio'] = Top15['Self-citations'] / Top15['Citations']
    country = Top15['Citation Ratio'].idxmax()
    max_ratio = Top15['Citation Ratio'].max()
    
    return (country, max_ratio)
```

```python
def answer_five():
    Top15 = answer_one()
    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    third_country = Top15['PopEst'].sort_values(ascending=False).index[2]
    
    return third_country
```

```python
def answer_six():
    Top15 = answer_one()
    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable docs per Capita'] = Top15['Citable documents'] / Top15['PopEst']
    
    corr = Top15['Citable docs per Capita'].astype(float).corr(Top15['Energy Supply per Capita'].astype(float))
    
    return corr
```

```python
answer_six()
```

```python
def answer_seven():
    Top15 = answer_one()
    
    ContinentDict  = {'China':'Asia', 
                      'United States':'North America', 
                      'Japan':'Asia', 
                      'United Kingdom':'Europe', 
                      'Russian Federation':'Europe', 
                      'Canada':'North America', 
                      'Germany':'Europe', 
                      'India':'Asia',
                      'France':'Europe', 
                      'South Korea':'Asia', 
                      'Italy':'Europe', 
                      'Spain':'Europe', 
                      'Iran':'Asia',
                      'Australia':'Australia', 
                      'Brazil':'South America'}
    
    Top15['PopEst'] = (Top15['Energy Supply'] / Top15['Energy Supply per Capita']).astype(float)
    
    grouped = Top15['PopEst'].groupby(ContinentDict).agg(['size', 'sum', 'mean', 'std'])
    grouped.index.name = 'Continent'
    
    return grouped
```
